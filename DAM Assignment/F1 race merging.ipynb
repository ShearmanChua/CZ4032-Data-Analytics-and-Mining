```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set()
```

```python
laps_df = pd.read_csv("C:/Users/Shearman Chua/Desktop/Com Sci Courses/Year 3 Semester 1/CZ4032-Data Analytics and Mining/Assignment/F1 data/lapTimes.csv")
print(laps_df.head(10))
```

```python
pitstops_df = pd.read_csv("C:/Users/Shearman Chua/Desktop/Com Sci Courses/Year 3 Semester 1/CZ4032-Data Analytics and Mining/Assignment/F1 data/pitStops.csv")
print(pitstops_df.head(10))
```

```python
laps_combined = pd.merge(laps_df, pitstops_df, how='left', on=['raceId', 'driverId','lap'])
print(laps_combined.head(20))
```

```python
laps_combined.rename(columns = {'time_x': 'lap_time', 'milliseconds_x':'milliseconds_lap','time_y': 'pit_time', 'duration':'pit_duration','milliseconds_y':'milliseconds_pit'}, inplace = True)
print(laps_combined.head(10))
```

```python
races_df = pd.read_csv("C:/Users/Shearman Chua/Desktop/Com Sci Courses/Year 3 Semester 1/CZ4032-Data Analytics and Mining/Assignment/F1 data/races.csv")
print(races_df.head(10))
```

```python
races_df.drop(['time', 'url'], axis='columns', inplace=True)
```

```python
laps_combined = pd.merge(races_df, laps_combined, on='raceId')
print(laps_combined.head(10))
```

```python
laps_combined = laps_combined.loc[laps_combined['year'] >= 2014]
```

```python
laps_combined.sort_values(by=['date','lap','position'], inplace=True)
laps_combined.head(10)
```

```python
driver_results_df = pd.read_csv("C:/Users/Shearman Chua/Desktop/Com Sci Courses/Year 3 Semester 1/CZ4032-Data Analytics and Mining/Assignment/F1 data/results.csv")
print(driver_results_df.head(10))
```

```python
driver_results_df.drop(['resultId'], axis='columns', inplace=True)
driver_results_df.rename(columns = {'number': 'car_no'}, inplace = True)
print(driver_results_df.head(10))
```

```python
driver_results_df.drop(['car_no','points'], axis='columns', inplace=True)
driver_results_df.rename(columns = {'rank': 'fastest_lap_rank','position':'final_position'}, inplace = True)
print(driver_results_df.head(10))
```

```python
laps_combined = pd.merge(laps_combined,driver_results_df, on=['raceId','driverId'])
print(laps_combined.head(20))
```

```python
laps_combined.sort_values(by=['date','lap','position'], inplace=True)
laps_combined.head(20)
```

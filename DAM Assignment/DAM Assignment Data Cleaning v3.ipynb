### Building Train and Test Dataset to aggregate data

```python
combined_df = combined_df.drop(['error_catagory','name_y','country','fullName', 'resultId'], axis=1)
combined_df.info()
```

```python
for index, row in combined_df.iterrows():
    if(row['downforce_level'] == 'High'):
        combined_df.at[index,'downforce_level'] = 3
    elif(row['downforce_level'] == 'Medium'):
        combined_df.at[index,'downforce_level'] = 2
    elif(row['downforce_level'] == 'Low'):
        combined_df.at[index,'downforce_level'] = 1
```

```python
combined_df['downforce_level'] = combined_df['downforce_level'].astype(str).astype(int)
combined_df.info()
```

```python
filtered_combined_no_change = combined_df
```

```python
grouped_df = filtered_combined_no_change.groupby(['driverId']) 
grouped_df.first()
```

```python
train_df = pd.DataFrame()
test_df = pd.DataFrame()

for name, group in grouped_df:
    group_size = len(group["constructorId"].unique())
    constructor_arr = group["constructorId"].unique()
    if(group_size == 1):
        train_df = train_df.append(group)
    else:
        for i in range(0,group_size):
            if(i < group_size-1):
                train_df = train_df.append(group[group["constructorId"] == constructor_arr[i]])
            else:
                test_df = test_df.append(group[group["constructorId"] == constructor_arr[i]])
```

```python
train_df_drop = train_df.dropna()
train_df_drop.info()
```

```python
test_df_drop = test_df.dropna()
test_df_drop.info()
```

```python
train_groupby = train_df_drop.groupby(['driverId', 'constructorId', 'year']).agg([np.mean])
train_groupby = train_groupby.reset_index()
train_groupby['year'] = train_groupby['year'].astype(int)
train_groupby.info()
```

```python
test_groupby = test_df_drop.groupby(['driverId', 'constructorId', 'year']).agg([np.mean])
test_groupby = test_groupby.reset_index()
test_groupby['year'] = test_groupby['year'].astype(int)
test_groupby.info()
```

```python
train_groupby.index = list(train_groupby.index)
train_groupby.info()
```

### Filter final standing

```python
results2_df = results2_df[(results2_df['year']>=2004) & (results2_df['year']<2018)]
results2_df.info()
```

```python
# Get last race for each season
results2_group = results2_df.groupby(['year']) 
last_race = pd.DataFrame()
for name, group in results2_group:
    last_race = last_race.append(group.tail(1))
last_race_final = last_race.drop(['circuitId', 'name'], axis=1)
last_race_final.head(20)
```

```python
# final results of each driver in each year from driverStandings
final_standing_df = pd.DataFrame()
for index, row in last_race_final.iterrows(): 
    temp_df = driverStandings_df[driverStandings_df['raceId']==row['raceId']]
    temp_df['year'] = row['year']
    temp_df['positionText'] = temp_df['positionText'].astype(int)
    temp_df = temp_df.sort_values(by=['positionText'])
    final_standing_df = final_standing_df.append(temp_df)
final_standing_df.info()    
final_standing_df[final_standing_df['raceId'] == 89]
```

```python
results_df.info()
```

```python
# get all results info with important attributes
results_Q3 = pd.merge(results_df, races_df, how='left', on=['raceId'])
results_Q3 = results_Q3.drop(['resultId', 'grid', 'positionOrder', 'points','laps','time','milliseconds','fastestLap','rank','fastestLapTime','fastestLapSpeed','statusId','name'], axis=1)
results_Q3.info()
```

```python
merged_results_standing = pd.merge(final_standing_df, results_Q3, how='left', on=['raceId', 'driverId','year'])
merged_results_standing.head(20)
```

```python
for index, row in merged_results_standing.iterrows():
    if(math.isnan(row['constructorId'])):
        temp = results_Q3[(results_Q3["driverId"] == row["driverId"]) & (results_Q3["year"] == row["year"])]
        constructor = temp['constructorId'].unique()[0]
        circuit = temp['circuitId'].unique()[0]
        merged_results_standing.at[index,'constructorId'] = constructor
        merged_results_standing.at[index,'circuitId'] = circuit
merged_results_standing.info()
```

```python
# merge both dataset in train dataset
for index, row in train_groupby.iterrows():
    
    temp = merged_results_standing[(merged_results_standing['driverId'] == row['driverId'].iloc[0]) & (merged_results_standing['constructorId'] == row['constructorId'].iloc[0]) & (merged_results_standing['year'] == row['year'].iloc[0])]
    if(temp['points'].empty == False):
        train_groupby.at[index,'points'] = temp['points'].iloc[0]
        train_groupby.at[index,'positionText'] = temp['positionText'].iloc[0]
        train_groupby.at[index,'wins'] = temp['wins'].iloc[0]
train_groupby.head()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set()
```

```python
main_df = pd.read_csv("Cleaned Data/dataset.csv")
print(main_df.head(10))
```

```python
races_df = pd.read_csv("Cleaned Data/races_cleaned.csv")
print(races_df.head(10))
```

```python
df_combined = pd.merge(races_df, main_df, on=['name','year'])
print(df_combined.head(10))
```

```python
drivers  = pd.read_csv("F1 data/drivers.csv")
drivers.head()
```

```python
df_combined = pd.merge(df_combined, drivers, on=['driverRef'])
df_combined.head()
```

```python
result = races_df = pd.read_csv("Cleaned Data/results_cleaned.csv")
result.drop(['statusId'], axis='columns', inplace=True)
result.head()
```

```python
df_combined = pd.merge(df_combined, result, on=['raceId','driverId'])
df_combined.drop(['position'], axis='columns', inplace=True)
df_combined.head()
```

```python
pos = []
# 0 is DNF
# 1 is podium
# 2 is 4 to 10
# 3 is finished
for index, row in df_combined.iterrows():
    if row['statusId'] == 0.0:
        pos.append(0)
    elif row['positionOrder'] < 4.0:
        pos.append(1)
    elif row['positionOrder'] >= 4.0 and row['positionOrder'] < 11.0:
        pos.append(2)
    else:
        pos.append(3)
```

```python
df_combined['race_class'] = pos
df_combined.head()
```

```python
combined  = pd.read_csv("./Output/question1 - final.csv")
combined.head()
```

```python
df_combined = pd.merge(df_combined, combined, on=['raceId','driverId','circuitId'])
df_combined.head()
```

```python
df_combined.drop(['raceId', 'year','name','driverRef','positionOrder','statusId','Super Soft','Ultra soft','Hard'], axis='columns', inplace=True)
df_combined.head()
```

```python
df_combined.drop(['Medium','Soft'], axis='columns', inplace=True)
df_combined.head()
```

```python
df_combined.to_csv('Data/question4.csv', index=False) 
```
